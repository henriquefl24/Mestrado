# Importação de Bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Arroz

## Estados

### Tabela de Dados

```python
tab = 'https://github.com/henriquefl24/Mestrado/blob/main/Arroz_Estados.xlsx?raw=true'

df = pd.read_excel(tab, 'Producao')
df = df.replace(('...','-'), np.NaN)
df = df.set_index('estados')
df.head(5)
```

### Limitando os dados para os ultimos 5 anos

```python
df['media'] = (df['2015'] + df['2016'] + df['2017'] + df['2018'] + df['2019'] + df['2020'])/6/10**3
df = df.sort_values(by='media', ascending=False)
df = df.loc[:'Maranhão', '2015':]
df.head(10)
```

# Gráficos

```python
sns.set(font_scale=1.6)
sns.set_style('white')
fig, axes = plt.subplots(figsize=(16, 12),
                         nrows=3,
                         ncols=2
                         )

# Graficos de Arroz e cidades

sns.barplot(ax=axes[0,0],
            data=df,
            x=df.media,
            y=df.index,
            palette='inferno')
axes[0,0].set_title('(A)')
axes[0,0].set_ylabel('Estados')
axes[0,0].set_xlabel('')

sns.barplot(ax=axes[1,0],
            data=df_RS,
            x=df_RS.media,
            y=df_RS.index,
            palette='Set2')
axes[1,0].set_title('(B)')
axes[1,0].set_ylabel('Municípios (RS)')
axes[1,0].set_xlabel('')

sns.barplot(ax=axes[2,0],
            data=df_SC,
            x=df_SC.media,
            y=df_SC.index,
            palette='flare')

axes[2,0].set_title('(C)')
axes[2,0].set_ylabel('Municípios (SC)')
axes[2,0].set_xlabel('Produção (mil tonelada)')

# Grafico de Cana-de-açúcar e cidades

sns.barplot(ax=axes[0,1],
            data=df_cana,
            x=df_cana.media,
            y=df_cana.index,
            palette='inferno')
axes[0,1].set_title('(D)')
axes[0,1].set_ylabel('')
axes[0,1].set_xlabel('')

sns.barplot(ax=axes[1,1],
            data=df_SP,
            x=df_SP.media,
            y=df_SP.index,
            palette='Set2')
axes[1,1].set_title('(E)')
axes[1,1].set_ylabel('Municípios (SP)')
axes[1,1].set_xlabel('')

sns.barplot(ax=axes[2,1],
            data=df_GO,
            x=df_GO.media,
            y=df_GO.index,
            palette='flare')

axes[2,1].set_title('(F)')
axes[2,1].set_ylabel('Municípios (GO)')
axes[2,1].set_xlabel('Produção (milhões de toneladas)')

plt.suptitle('Produção média de arroz e cana-de-açúcar (2015-2020)')
fig.tight_layout()
plt.savefig('Critério', dpi=400, tranparent=True, bbox_inches='tight')

plt.plot();
```
